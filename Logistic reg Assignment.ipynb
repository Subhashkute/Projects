# Whether the client has subscribed a term deposit or not 


```python
import pandas as pd
```

```python
bank= pd.read_csv('bank-full.csv', sep=';')
bank
```

```python
import pandas_profiling as pp
```

```python
Bank=pp.ProfileReport(bank)
Bank.to_file(output_file="Bank.html")
```

```python
bank.describe()
```

```python
bank1=pd.get_dummies(bank,columns=['housing','loan','y','default','job','marital','education','contact','poutcome','month'])
bank1
```

```python
bank2 =bank1.drop(['housing_no','loan_no','y_no','default_no'],axis=1)
bank2
```

```python
x=bank2.drop('y_yes',axis=1)
```

```python
y=bank2['y_yes']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size = 0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model =  LogisticRegression(fit_intercept = True)

model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
ypred
```

```python
x_test.shape
```

```python
model.score(x_test,y_test)
```

As we can see accuracy of our model is 0.8863 , we can predict whether the client has subscribed a term deposit or not
