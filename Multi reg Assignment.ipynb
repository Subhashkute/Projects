# Prediction model for profit of 50_startups data.

```python
import pandas as pd
```

```python
startups = pd.read_csv('50_Startups.csv')
```

```python
startups
```

```python
startup1=startups.drop('State',axis=1)
startup1
```

```python
startup2=startup1.rename({'R&D Spend':"RD",'Marketing Spend':'Marketing_Spend'},axis=1)
startup2
```

```python
import seaborn as sns
sns.pairplot(startup2)
```

```python
import statsmodels.formula.api as smf
SUP = smf.ols('Profit~ RD + Administration + Marketing_Spend',data=startup2).fit()
```

```python
SUP.summary()
```

Simple linear regression

```python
SUP_m1=smf.ols("Profit~Administration",data=startup2).fit()
```

```python
SUP_m1.summary()
```

```python
SUP_m2=smf.ols("Profit~Marketing_Spend",data=startup2).fit()
```

```python
SUP_m2.summary()
```

```python
SUP_m3=smf.ols("Profit~Administration+Marketing_Spend",data=startup2).fit()
```

```python
SUP_m3.summary()
```

Calculating VIF

```python
r_sq_RD =smf.ols('RD~Administration+Marketing_Spend',data=startup2).fit().rsquared
vif_RD = 1/(1-r_sq_RD)

r_sq_ad =smf.ols('Administration~RD+Marketing_Spend',data=startup2).fit().rsquared
vif_ad = 1/(1-r_sq_ad)

r_sq_ms =smf.ols('Marketing_Spend~Administration+RD',data=startup2).fit().rsquared
vif_ms = 1/(1-r_sq_ms)

d={'Var':['RD','ADM',"MS"],'VIF':[vif_RD,vif_ad,vif_ms]}
VIF=pd.DataFrame(d)
VIF

# RD and Marketing spend show high collinearity
```

Residual analysis

To test for normality we use qqplot

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
qqplot=sm.qqplot(SUP.resid,line='q')
plt.show()
```

As majority of points lie on trend line we can say residual follows normal dist

Homoscedasticity - variance is constant for each independent variable

```python
def get_std_val(val):
    return (val-val.mean())/val.std()
```

```python
plt.scatter(get_std_val(SUP.fittedvalues),get_std_val(SUP.resid))

plt.title("Residual Plot")
plt.show()
```

As the plot shows no trend this means our model is good

Residual VS Regressor

```python
from statsmodels.graphics.regressionplots import influence_plot
```

```python
fig=plt.figure(figsize=(15,8))
fig= sm.graphics.plot_regress_exog(SUP,"RD",fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
fig= sm.graphics.plot_regress_exog(SUP,"Administration",fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
fig= sm.graphics.plot_regress_exog(SUP,"Marketing_Spend",fig=fig)
plt.show()
```

Outlier Detection

```python
influence_plot(SUP)
plt.show()
```

```python
k=startup2.shape[1]
n=startup2.shape[0]

leverage_cutoff =3*((k+1)/n)
leverage_cutoff
```

As we do not have any point crossing leverage cutoff i.e 0.3 we conclude that there is no outlier present  in the data
