```python
# normalizing the data
from sklearn.preprocessing import scale

df_normal=scale(wine)
df_normal
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=3)
pca_values=pca.fit_transform(df_normal)
pca_values
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
a=(pca_values[:,0:3])
a
```

```python
new_df=pd.DataFrame(a,columns=['pc1','pc2','pc3'])
new_df
```

```python
import seaborn as sns
sns.scatterplot(data=new_df,x='pc1',y='pc2',hue='pc3')
```

# Hierarchial clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering 
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,8))
dg= sch.dendrogram(sch.linkage(df_normal,method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
```

```python
y_hc = hc.fit_predict(df_normal)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
wine['h_clusterid'] = Clusters
wine
```

```python
wine['h_clusterid'].value_counts()
```

# K-Means

```python
from sklearn.cluster import KMeans
```

```python
l=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(df_normal)
    l.append(kmeans.inertia_)
plt.plot(range(1, 11),l)
```

```python
kmodel=KMeans(n_clusters=3).fit(df_normal)
kmodel
```

```python
kmodel.labels_
```
