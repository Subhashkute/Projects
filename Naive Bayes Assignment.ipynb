```python
import pandas as pd
```

```python
train=pd.read_csv('SalaryData_Train(1).csv')
train
```

```python
test=pd.read_csv('SalaryData_Test(1).csv')
```

```python
train.describe()
```

```python
train.info()
```

```python
train['educationno'].value_counts()
```

```python
string_col = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
for i in string_col:
    train[i]= labelencoder.fit_transform( train[i])
    test[i]=labelencoder.fit_transform(test[i])
```

```python
train.head()
```

```python
mapping = {' >50K': 1, ' <=50K': 2}
```

```python
train = train.replace({'Salary': mapping})
test = test.replace({'Salary': mapping})
```

```python
x_train = train.iloc[:,0:13]
y_train = train.iloc[:,13]
x_test = test.iloc[:,0:13]
y_test = test.iloc[:,13]
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
classifier_mb = MB()
```

```python
classifier_mb.fit(x_train,y_train)
```

```python
y_pred_mb=classifier_mb.predict(x_test)
```

```python
import numpy as np
print ("Accuracy",np.mean(y_pred_mb==y_test.values.flatten()))
```

```python
from sklearn.naive_bayes import GaussianNB
classifier_gb = GaussianNB()
```

```python
classifier_gb.fit(x_train,y_train)
```

```python
y_pred_gb=classifier_gb.predict(x_test)
```

```python
print ("Accuracy",np.mean(y_pred_gb==y_test.values.flatten()))
```
